Tenemos los datos:

El púlsar se describe con 8 variables que vienen de dos señales distintas: el **perfil integrado** (la forma promedio del pulso 
de radio) y la **curva DM-SNR** (relacionada con la dispersión de la señal al 
atravesar el medio interestelar).

El objetivo de esta descripcción es hacerse una idea de qué representa la variable, sin más. 

- Profile_mean: altura promedio de la señal 
- Profile_stdev: cuánto varía la señal alrededor de su media
- Profile_skewness: asimetría de la forma de la señal
- Profile_kurtosis: qué tan afilado es el pico 

- DM_mean: valor promedio de la curva de dispersión.
- DM_stdev: variabilidad de esa curva alrededor de su media.
- DM_skewness: asimetría de la curva DM-SNR.
- DM_kurtosis: qué tan afilado es el pico de la curva DM-SNR.

Un aspecto característico de un púlsar es que emite una señal con un pico bien definido (un valor alto de "Profile_kurtosis").

```python
import pandas as pd 
from ucimlrepo import fetch_ucirepo

pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)

htru2 = fetch_ucirepo(id=372)
X = htru2.data.features
y = htru2.data.targets

datos = pd.concat([X, y],axis=1)
datos.groupby('class')

datos.groupby("class").mean()
```

## Visualización de los datos

Naranja: Púlsar
Azul: No púlsar


```python
import matplotlib.pyplot as plt

plt.hist(datos[datos["class"]==0]["Profile_kurtosis"], label = "No púlsar")
plt.hist(datos[datos["class"]==1]["Profile_kurtosis"], label = "Púlsar")
plt.grid()
plt.legend()
plt.show()
```

En el útlimo gráfico, se ve claramente que los púlsars tienen mucha más kurtosis.

```python
plt.hist(datos[datos["class"]==0]["Profile_skewness"], label = "No púlsar")
plt.hist(datos[datos["class"]==1]["Profile_skewness"], label = "Púlsar")
plt.grid()
plt.legend()
plt.show()
```

Ahora Profile_skewness normalizada:

```python
plt.hist(datos[datos["class"]==0]["Profile_skewness"],density = True, label = "No púlsar")
plt.hist(datos[datos["class"]==1]["Profile_skewness"], density = True, label = "Púlsar")
plt.grid()
plt.legend()
plt.show()
```
